Here is a breakdown of some of the audio features that are present in the audio features dataset:

1. Danceability: A measurement criteria combining musical elements in terms of suitability for dancing. (Numeric)
2. Energy: Characterizes the intensity and activity within each recording. (Numeric)
3. Key: Showcases tonalities such as C major or D minor described using text representations.
4. Loudness: Expressed as decibel levels measuring the overall volume across songs. (Numeric)
5. Mode: Distinguishes major or minor tonalities indicated as numeric values within this context. (Numeric)
6. Speechiness: Shows the presence of spoken-word elements quantitatively within tracks. (Numeric)
7. Acousticness: Displays the acoustic qualities of songs numerically, reflecting the contrast between natural sound and electronically enhanced production or elements. (Numeric)
8. Instrumentalness: Quantifies the likelihood of a song being instrumental using numeric metrics. (Numeric)
9. Liveness: Reveals the presence of an audience within live recordings via numerical evaluation. (Numeric)
10. Valence: Describes the musical positivity conveyed by songs quantitatively as numeric measurements on a scale. (Numeric)
11. Tempo: Measures the beats per minute (BPM) for tempo indication evaluated by numeric means. (Numeric)
12. Time Signature: Specifies the song structure like 4/4 or 3/4 using text representation. 
13. Spotify Track Popularity: Gauges the popularity of the song on Spotify. (Numeric)

```python
df_numeric
```

I can drop the additional columns that are not variables of the audio features, such as which 'WeekID, 'Weeks on Chart', 'Week Position', and 'Peak Position'. After dropping these columns I am left with just the numeric audio features. 

I also group the df_popularity_features dataframe by 'SongID' which eliminates having the same songID appear in several rows. I can use the .mean() after grouping by 'SongID' without altering the data since all the audio features for the same SongID entries are all the same. 

```python
df_popularity_features = df_numeric.drop(['WeekID','Instance', 'Week Position', 'Peak Position', 'Weeks on Chart'], axis = 1)
df_popularity_features = df_popularity_features.groupby('SongID').mean()
df_popularity_features
```

```python
audio_features = ['spotify_track_duration_ms', 'danceability', 'energy', 'key','loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness','liveness', 'valence', 'tempo', 'time_signature']
sns.pairplot(data = df_popularity_features,y_vars = 'spotify_track_popularity', x_vars = audio_features)
plt.suptitle("Spotify Popularity vs. Other Audio Features")
```

From the pairplot() function, I plot all the audio features on the x-axis vs. spotify_track_popularity on the y-axis to see if I can spot any correlations. From first glance, it is hard to discern if the there are any correlations with the popularity variable. With a closer look, it appears as though the loudness and danceability variables might have some slight correlation with the spotify_track_popularity variable as it does appear they slightly increase with popularity. 

For a more quantitative investigation I can calculate the correlation coefficients of each variable.

```python
df_popularity_features.corr()
```

Using a heat map to visualize the differences in the correlation coefficient is a quick to spot which one is higher without having to manually look and compare each number of the correlation coefficients. The darker colors are associated with a lower correlation and and lighter colors are associated with a higher correlation. 

```python
plt.figure(figsize = (15, 15))
sns.heatmap(df_popularity_features.corr(), annot = True, fmt = ".4f")
plt.title("Correlation Coefficients across each Variable")
```

The heatmap along with the correlation coefficients provide me with much more visual insight than just the scattergraphs. From the spotify_track_popularity variable row of the correlation coefficient heatmap, I can see that the 'loudness' variable visually has the lightest heatmap color and therefore is the most postively correlated with spotify_track_popularity. Loudness is followed by speechiness, which has the second lightest heatmap color. Thirdly and fourthly are the variables spotify_track_duration_ms and danceability. Loudness, speechiness, spotify_track_duration_ms, and danceability have correlation coefficients of .3484, .2116., .1856, and .1809, respectively. This shows me that as these positively correlated variables increase, spotify_track_popularity also increases. This implies that the variables of loudness, speechiness, longer duration of songs, and danceability are generally enjoyed by the general public in their songs.

Interestingly, the acousticness variable has the stronges negative correlation with popularity with a correlation coefficient of -.3060, as visually shown through the darker heatmap color. The valence variable has the next most negative correlations with popularity with a coefficient of -.2178. As these negatively correlated variables increase, spotify_track_popularity decreases. This implies that these variables of acousticness, and valance are not generally enjoyed by the public.

## Research Question #2: Can how many weeks the song stays on the chart and if the song becomes a top 10 hit be predicted by it's audio features?

#### Analysis

##### First, I want to explore if I can use the audio features of each song to predict how long the song will remain on the chart for. The dataset has a  "Weeks on Chart" variable that lists many weeks each song has been on the list at the date time point at which the entry is recorded. Since this "weeks on chart" variable is already present in the dataframe, I can use a supervised machine learning algorithm such as linear regression. Exploring the number of weeks the song will be on the list utilizes continuous numerical outputs, which is another reason why linear regression is a good algorithm to excercise here.

In this dataset, there are multiple instances of the same song as each song can be in the top 100 list at different time points. Therefore, I use the groupby() function to group the data by the 'SongID', which contains both the song name and the artist. I then use the .max() function to see what the max weeks on chart is for each individual SongID since I am only interested in the max value of number of weeks on chart for each song. 

```python
df_weeks_on_chart = df_filtered.groupby(['SongID']).max()
df_weeks_on_chart
```

Plotting the distribution of 'Weeks on Chart':

```python
sns.set(style = 'whitegrid')
f, ax = plt.subplots(1,1, figsize = (8,5))
ax = sns.histplot(df_weeks_on_chart['Weeks on Chart'], kde =True, color = 'c')
```

I will now drop the categorical or discrete columns that are not necessary for this linear regression analysis. 

```python
df_numeric_weeks_on_chart = df_weeks_on_chart.drop(['index_x', 'WeekID', 'Week Position', 'Instance', 'Peak Position', 'spotify_genre', 'spotify_track_explicit', 'time_signature', 'mode', 'key'], axis = 1)
df_numeric_weeks_on_chart
```

Creating the dataframe X_data_weeks where the target label 'Weeks on Chart' is removed and stored in a different dataframe, y_target_weeks. The dataframe X_data_weeks contains only the continuous numeric variables/audio features that I want to look at. 

```python
X_data_weeks = df_numeric_weeks_on_chart.drop('Weeks on Chart', axis = 1)
y_target_weeks = df_numeric_weeks_on_chart['Weeks on Chart']
```

```python
X_data_weeks
```

I am shuffling the data before splitting it into training and testing sets. 

```python
X_data_weeks, y_target_weeks = shuffle(X_data_weeks, y_target_weeks)
```

Here I use the train_test_split method to split the data into 80% train set and 20% test set. 

```python
X_train, X_test, y_train, y_test = train_test_split(X_data_weeks, y_target_weeks, test_size = 0.2, random_state = 5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

We can see the sizes of each of the data sets and that the training sets are much larger than the testing sets.

I use the StandardScaler() from sklearn to scale all the variables before creating the linear regression model and fitting the it to the training sets. 

```python
scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns, index = X_train.index)
X_test = pd.DataFrame(scaler.fit_transform(X_test), columns = X_test.columns, index = X_test.index)

regression = LinearRegression()
regression.fit(X_train, y_train)
```

Now that the linear regression model is created, I use it to predict on the test set. Once it is predicted, I print out the intercept and coefficients for each input feature. I also print out the root mean squred error to see the standard error associated with the model.

```python
linreg_predictions = regression.predict(X_test)
```

```python
print("The intercept is: ", regression.intercept_)
print("The coefficients for each input feature are: ", regression.coef_)
```

```python
rmse = math.sqrt(mean_squared_error(linreg_predictions, y_test))
print("The root mean squared error is: ", rmse)
```

The linear regression model trained by the training set was able to predict on the test set with a root mean squared value of 7.3489. By looking at the statistics of the X_test dataset with the describe() function below, a root mean squared value of 7.3 is relatively large compared to the range of the columns. Another statistic we can use to see the accuracy of our model is to use the r2 value, which I calculate below. 

```python
X_test.describe()
```

```python
r2_score(linreg_predictions, y_test)
```

Since the r2 value is negative, this shows that the model does not fit the data at all and that it has no predictive value. This may be because of using the wrong model or I should have removed some of the other variables also to get a better fit. Either way, this shows that using the variables that I did, I was not able to predict the number of weeks on the charts using linear regression. 

I can see from the coefficients that the 'spotify_track_popularity' feature has the highest positive coefficient when compared to the other audio features. When I visualize this feature in the graph below, I can see that the X_test data for the 'spotify_track_popularity' resembles more of a positively sloped line than many of the other columns, but the model is still unable to predict. Here, I visually plot the discrepency in predictions vs actual. 

```python
plt.scatter(X_test.iloc[:,10], y_test, color = 'black', label = 'Actual')
plt.plot(X_test.iloc[:,10], linreg_predictions, color = 'red', label = 'Predicted')
plt.xlabel('Spotify_track_popularity Feature')
plt.ylabel('Weeks on Chart')
plt.title('Actual vs. Predicted for Spotify track popularity feature')
plt.legend()
plt.show()
```

## Research Question #3: How have explicit songs in the Billboard Hot 100 changed in popularity and prevalence throughout the years?

#### Analysis

With this researh question, I want to explore how explicit songs have changed over the course of the Billboard Hot list in terms of how many songs are on the list and also the average popularity of these explicit songs. First, I will explore the number of explicit songs in the list for each year.

I create a new dataframe df_explicit that changes the 'spotify_track_explicit' column from true/false object values to int values (1 for true, 0 for false). I then update df_explicit to only contain explicit songs.

```python
df_explicit = df_filtered.copy()
df_explicit['spotify_track_explicit'] = df_explicit['spotify_track_explicit'].astype(int)
df_explicit = df_explicit[df_explicit['spotify_track_explicit'] == 1]
```

```python
df_explicit
```

In order to sum up the number of explicit songs on the list per year, I need to create a new column 'year' that returns only the year the song is listed on the Billboard Hot 100 list. I can then group the dataframe by year and sum up the total number of explicit songs by summing up the 'spotify_track_explicit column since each explicit song has a value of '1' in that column. 

This new array total_explicit_yearly shows me the total number of explicit songs in the Billboard Hot 100 list throughout the year. 

```python
df_explicit['year'] = pd.to_datetime(df_explicit['WeekID']).dt.year
total_explicit_yearly = df_explicit.groupby('year')['spotify_track_explicit'].sum()
total_explicit_yearly
```

```python
plt.figure(figsize=(15, 6))
plt.plot(total_explicit_yearly.index, total_explicit_yearly.values, marker='o', linestyle='-')
plt.title('Total Instances of Explicit Songs in Billboard 100 Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

When this is graphed, I can see that number of explicit songs in the Billboard Hot 100 list increases throughout the years. Up throughout 1990, there are not many songs on the list at all, but after 1990 the amount of songs on the list throughout the year inreases dramatically. There is another huge spike in the mid 2010s. There appears to be a drop off in 2021, and I can explore as to why this is. 

I can use slicing to select just the songs from 2021 that were on the list and are explicit and store in the dataframe songs_2021_explicit. From songs_2021_explicit I can see that the last WeekID entry is only May 01, 2021. Therefore, I know the reason why 2021 had a dropoff in explicit songs is because the dataframe ends halfway through 2021. 

From this new dataframe I can also see that since there is a time component to this dataframe, the same explicit song can appear on the list several times at different time points. This may skew the data because we could end up having the same song appear in the Billboard Hot 100 many weeks of the year and it would appear as if there were many explicit songs to be on the list that year, but it would simply be the same song appearing many times. I can do some more exploring to try and see how many unique explicit songs are on the Billboard Hot 100 list per year to see if this changes our analysis at all.
